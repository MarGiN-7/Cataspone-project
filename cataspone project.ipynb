```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.metrics import accuracy_score, precision_score, recall_score, roc_auc_score, confusion_matrix, classification_report
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.cluster import KMeans

import warnings
warnings.filterwarnings('ignore')

print("PART 1: CUSTOMER CHURN PREDICTION")
print("="*50)

churn_url = "https://raw.githubusercontent.com/IBM/telco-customer-churn-on-icp4d/master/data/Telco-Customer-Churn.csv"
churn_df = pd.read_csv(churn_url)

print("\nDataset Shape:", churn_df.shape)
print("\nColumn Types:")
print(churn_df.dtypes)
print("\nFirst 5 Rows:")
print(churn_df.head())

print("\nMissing Values:")
print(churn_df.isnull().sum())

print("\nBasic Statistics:")
print(churn_df.describe())

print("\nChurn Distribution:")
print(churn_df['Churn'].value_counts())
print(churn_df['Churn'].value_counts(normalize=True))

plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
churn_df['Churn'].value_counts().plot(kind='bar')
plt.title('Churn Distribution')
plt.xlabel('Churn')
plt.ylabel('Count')

plt.subplot(1, 2, 2)
churn_df['Churn'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title('Churn Percentage')
plt.tight_layout()
plt.show()

plt.figure(figsize=(15, 10))
plt.subplot(2, 2, 1)
sns.countplot(data=churn_df, x='gender', hue='Churn')
plt.title('Churn by Gender')

plt.subplot(2, 2, 2)
sns.countplot(data=churn_df, x='Contract', hue='Churn')
plt.title('Churn by Contract Type')

plt.subplot(2, 2, 3)
sns.countplot(data=churn_df, x='InternetService', hue='Churn')
plt.title('Churn by Internet Service')

plt.subplot(2, 2, 4)
sns.countplot(data=churn_df, x='PaymentMethod', hue='Churn')
plt.xticks(rotation=45)
plt.title('Churn by Payment Method')
plt.tight_layout()
plt.show()

churn_df['TotalCharges'] = pd.to_numeric(churn_df['TotalCharges'], errors='coerce')

print("\nMissing Values After Conversion:")
print(churn_df.isnull().sum())

churn_df['TotalCharges'].fillna(churn_df['TotalCharges'].median(), inplace=True)

numerical_cols = ['tenure', 'MonthlyCharges', 'TotalCharges']

plt.figure(figsize=(15, 4))
for i, col in enumerate(numerical_cols):
    plt.subplot(1, 3, i+1)
    sns.boxplot(data=churn_df, y=col)
    plt.title(f'{col} Boxplot')
plt.tight_layout()
plt.show()

for col in numerical_cols:
    Q1 = churn_df[col].quantile(0.25)
    Q3 = churn_df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    churn_df[col] = churn_df[col].clip(lower_bound, upper_bound)

churn_processed = churn_df.copy()
churn_processed.drop('customerID', axis=1, inplace=True)

le = LabelEncoder()
categorical_cols = churn_processed.select_dtypes(include='object').columns

for col in categorical_cols:
    churn_processed[col] = le.fit_transform(churn_processed[col])

print("\nProcessed Data Types:")
print(churn_processed.dtypes)

X_churn = churn_processed.drop('Churn', axis=1)
y_churn = churn_processed['Churn']

X_train_churn, X_test_churn, y_train_churn, y_test_churn = train_test_split(
    X_churn, y_churn, test_size=0.2, random_state=42, stratify=y_churn
)

print("\nTraining Set Size:", X_train_churn.shape[0])
print("Test Set Size:", X_test_churn.shape[0])

scaler = StandardScaler()
X_train_churn_scaled = scaler.fit_transform(X_train_churn)
X_test_churn_scaled = scaler.transform(X_test_churn)

lr_model = LogisticRegression(max_iter=1000, random_state=42)
lr_model.fit(X_train_churn_scaled, y_train_churn)

rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
rf_model.fit(X_train_churn, y_train_churn)

print("\nLogistic Regression Results:")
lr_preds = lr_model.predict(X_test_churn_scaled)
lr_probs = lr_model.predict_proba(X_test_churn_scaled)[:, 1]
print("Accuracy:", accuracy_score(y_test_churn, lr_preds))
print("Precision:", precision_score(y_test_churn, lr_preds))
print("Recall:", recall_score(y_test_churn, lr_preds))
print("ROC-AUC:", roc_auc_score(y_test_churn, lr_probs))
print("\nClassification Report:")
print(classification_report(y_test_churn, lr_preds))

print("\nRandom Forest Results:")
rf_preds = rf_model.predict(X_test_churn)
rf_probs = rf_model.predict_proba(X_test_churn)[:, 1]
print("Accuracy:", accuracy_score(y_test_churn, rf_preds))
print("Precision:", precision_score(y_test_churn, rf_preds))
print("Recall:", recall_score(y_test_churn, rf_preds))
print("ROC-AUC:", roc_auc_score(y_test_churn, rf_probs))
print("\nClassification Report:")
print(classification_report(y_test_churn, rf_preds))

plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
sns.heatmap(confusion_matrix(y_test_churn, lr_preds), annot=True, fmt='d', cmap='Blues')
plt.title('Logistic Regression Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('Actual')

plt.subplot(1, 2, 2)
sns.heatmap(confusion_matrix(y_test_churn, rf_preds), annot=True, fmt='d', cmap='Blues')
plt.title('Random Forest Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.tight_layout()
plt.show()

feature_importance = pd.DataFrame({
    'Feature': X_churn.columns,
    'Importance': rf_model.feature_importances_
}).sort_values('Importance', ascending=False)

print("\nKey Churn-Driving Factors:")
print(feature_importance.head(10))

plt.figure(figsize=(10, 6))
sns.barplot(data=feature_importance.head(10), x='Importance', y='Feature')
plt.title('Top 10 Churn-Driving Factors')
plt.tight_layout()
plt.show()

print("\n" + "="*50)
print("PART 2: CREDIT RISK ANALYSIS")
print("="*50)

credit_url = "https://raw.githubusercontent.com/dphi-official/Datasets/master/Loan_Data/loan_train.csv"
credit_df = pd.read_csv(credit_url)

print("\nDataset Shape:", credit_df.shape)
print("\nColumn Types:")
print(credit_df.dtypes)
print("\nFirst 5 Rows:")
print(credit_df.head())
print("\nDataset Info:")
print(credit_df.info())

print("\nMissing Values:")
print(credit_df.isnull().sum())
print("\nMissing Value Percentage:")
print((credit_df.isnull().sum() / len(credit_df)) * 100)

print("\nBasic Statistics:")
print(credit_df.describe())

print("\nLoan Status Distribution:")
print(credit_df['Loan_Status'].value_counts())

numerical_credit = credit_df.select_dtypes(include=['int64', 'float64']).columns
categorical_credit = credit_df.select_dtypes(include='object').columns

for col in numerical_credit:
    credit_df[col].fillna(credit_df[col].median(), inplace=True)

for col in categorical_credit:
    credit_df[col].fillna(credit_df[col].mode()[0], inplace=True)

print("\nMissing Values After Cleaning:")
print(credit_df.isnull().sum())

plt.figure(figsize=(15, 10))
for i, col in enumerate(['ApplicantIncome', 'CoapplicantIncome', 'LoanAmount', 'Loan_Amount_Term']):
    plt.subplot(2, 2, i+1)
    sns.boxplot(data=credit_df, y=col)
    plt.title(f'{col} Distribution')
plt.tight_layout()
plt.show()

for col in ['ApplicantIncome', 'CoapplicantIncome', 'LoanAmount']:
    Q1 = credit_df[col].quantile(0.25)
    Q3 = credit_df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    credit_df[col] = credit_df[col].clip(lower, upper)

credit_df['Total_Income'] = credit_df['ApplicantIncome'] + credit_df['CoapplicantIncome']
credit_df['Income_Loan_Ratio'] = credit_df['Total_Income'] / (credit_df['LoanAmount'] + 1)
credit_df['EMI'] = credit_df['LoanAmount'] / (credit_df['Loan_Amount_Term'] + 1)
credit_df['Balance_Income'] = credit_df['Total_Income'] - (credit_df['EMI'] * 1000)
credit_df['LoanAmount_Log'] = np.log1p(credit_df['LoanAmount'])
credit_df['Total_Income_Log'] = np.log1p(credit_df['Total_Income'])

credit_df.replace([np.inf, -np.inf], 0, inplace=True)

print("\nNew Features Created:")
print(credit_df[['Total_Income', 'Income_Loan_Ratio', 'EMI', 'Balance_Income']].head())

credit_processed = credit_df.copy()
credit_processed.drop('Loan_ID', axis=1, inplace=True)

le_credit = LabelEncoder()
for col in credit_processed.select_dtypes(include='object').columns:
    credit_processed[col] = le_credit.fit_transform(credit_processed[col])

X_credit = credit_processed.drop('Loan_Status', axis=1)
y_credit = credit_processed['Loan_Status']

X_train_credit, X_test_credit, y_train_credit, y_test_credit = train_test_split(
    X_credit, y_credit, test_size=0.2, random_state=42, stratify=y_credit
)

print("\nTraining Set Size:", X_train_credit.shape[0])
print("Test Set Size:", X_test_credit.shape[0])

scaler_credit = StandardScaler()
X_train_credit_scaled = scaler_credit.fit_transform(X_train_credit)
X_test_credit_scaled = scaler_credit.transform(X_test_credit)

lr_credit = LogisticRegression(max_iter=1000, random_state=42)
lr_credit.fit(X_train_credit_scaled, y_train_credit)

rf_credit = RandomForestClassifier(n_estimators=100, random_state=42)
rf_credit.fit(X_train_credit, y_train_credit)

print("\nLogistic Regression - Credit Risk:")
lr_credit_preds = lr_credit.predict(X_test_credit_scaled)
lr_credit_probs = lr_credit.predict_proba(X_test_credit_scaled)[:, 1]

print("\nConfusion Matrix:")
print(confusion_matrix(y_test_credit, lr_credit_preds))
print("\nAUC Score:", roc_auc_score(y_test_credit, lr_credit_probs))
print("\nClassification Report:")
print(classification_report(y_test_credit, lr_credit_preds))

print("\nRandom Forest - Credit Risk:")
rf_credit_preds = rf_credit.predict(X_test_credit)
rf_credit_probs = rf_credit.predict_proba(X_test_credit)[:, 1]

print("\nConfusion Matrix:")
print(confusion_matrix(y_test_credit, rf_credit_preds))
print("\nAUC Score:", roc_auc_score(y_test_credit, rf_credit_probs))
print("\nClassification Report:")
print(classification_report(y_test_credit, rf_credit_preds))

plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
sns.heatmap(confusion_matrix(y_test_credit, lr_credit_preds), annot=True, fmt='d', cmap='Greens')
plt.title('Logistic Regression - Credit Risk')
plt.xlabel('Predicted')
plt.ylabel('Actual')

plt.subplot(1, 2, 2)
sns.heatmap(confusion_matrix(y_test_credit, rf_credit_preds), annot=True, fmt='d', cmap='Greens')
plt.title('Random Forest - Credit Risk')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.tight_layout()
plt.show()

X_credit_full_scaled = scaler_credit.transform(X_credit)
credit_df['Risk_Score'] = lr_credit.predict_proba(X_credit_full_scaled)[:, 1]

credit_df['Risk_Category'] = pd.cut(
    credit_df['Risk_Score'],
    bins=[0, 0.3, 0.6, 1.0],
    labels=['Low Risk', 'Medium Risk', 'High Risk']
)

print("\nRisk Score Distribution:")
print(credit_df['Risk_Category'].value_counts())

print("\nSample Risk Scores:")
print(credit_df[['Loan_ID', 'ApplicantIncome', 'LoanAmount', 'Credit_History', 'Risk_Score', 'Risk_Category']].head(10))

plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
sns.histplot(credit_df['Risk_Score'], bins=20, kde=True)
plt.title('Risk Score Distribution')

plt.subplot(1, 2, 2)
credit_df['Risk_Category'].value_counts().plot(kind='bar')
plt.title('Risk Category Distribution')
plt.tight_layout()
plt.show()

print("\n" + "="*50)
print("PART 3: MARKETING CAMPAIGN EFFECTIVENESS ANALYSIS")
print("="*50)

marketing_url = "https://raw.githubusercontent.com/nailson/ifood-data-business-analyst-test/master/ifood_df.csv"
marketing_df = pd.read_csv(marketing_url)

print("\nDataset Shape:", marketing_df.shape)
print("\nColumn Names:")
print(marketing_df.columns.tolist())
print("\nFirst 5 Rows:")
print(marketing_df.head())

print("\nBasic Statistics:")
print(marketing_df.describe())

print("\nMissing Values:")
print(marketing_df.isnull().sum())

marketing_df.dropna(inplace=True)

campaign_cols = [col for col in marketing_df.columns if 'Accepted' in col or 'Cmp' in col]
print("\nCampaign Columns Found:", campaign_cols)

product_cols = [col for col in marketing_df.columns if 'Mnt' in col]
print("Product Spending Columns:", product_cols)

purchase_cols = [col for col in marketing_df.columns if 'Purchases' in col or 'NumWeb' in col or 'NumCatalog' in col or 'NumStore' in col]
print("Purchase Columns:", purchase_cols)

plt.figure(figsize=(15, 10))

plt.subplot(2, 2, 1)
if 'Income' in marketing_df.columns:
    sns.histplot(marketing_df['Income'], bins=30, kde=True)
    plt.title('Income Distribution')

plt.subplot(2, 2, 2)
if len(product_cols) > 0:
    marketing_df[product_cols].sum().plot(kind='bar')
    plt.title('Total Spending by Product Category')
    plt.xticks(rotation=45)

plt.subplot(2, 2, 3)
if len(campaign_cols) > 0:
    campaign_acceptance = marketing_df[campaign_cols].sum()
    campaign_acceptance.plot(kind='bar', color='green')
    plt.title('Campaign Acceptance Count')
    plt.xticks(rotation=45)

plt.subplot(2, 2, 4)
if 'Education' in marketing_df.columns:
    marketing_df['Education'].value_counts().plot(kind='bar')
    plt.title('Customer Education Distribution')
    plt.xticks(rotation=45)

plt.tight_layout()
plt.show()

print("\nCampaign Acceptance Rates:")
if len(campaign_cols) > 0:
    print(marketing_df[campaign_cols].mean())

if 'Income' in marketing_df.columns:
    marketing_df['Income'].fillna(marketing_df['Income'].median(), inplace=True)
    
segment_features = ['Income']
if len(product_cols) > 0:
    marketing_df['Total_Spending'] = marketing_df[product_cols].sum(axis=1)
    segment_features.append('Total_Spending')

segment_data = marketing_df[segment_features].copy()
segment_data.fillna(segment_data.median(), inplace=True)

scaler_seg = StandardScaler()
segment_scaled = scaler_seg.fit_transform(segment_data)

kmeans = KMeans(n_clusters=4, random_state=42, n_init=10)
marketing_df['Customer_Segment'] = kmeans.fit_predict(segment_scaled)

print("\nCustomer Segments:")
print(marketing_df['Customer_Segment'].value_counts())

print("\nSegment Characteristics:")
segment_analysis = marketing_df.groupby('Customer_Segment').agg({
    'Income': 'mean',
    'Total_Spending': 'mean' if 'Total_Spending' in marketing_df.columns else 'count'
}).round(2)
print(segment_analysis)

plt.figure(figsize=(10, 6))
if 'Total_Spending' in marketing_df.columns:
    sns.scatterplot(data=marketing_df, x='Income', y='Total_Spending', hue='Customer_Segment', palette='viridis')
    plt.title('Customer Segmentation')
    plt.tight_layout()
    plt.show()

print("\nCampaign Performance Metrics:")
if len(campaign_cols) > 0:
    for col in campaign_cols:
        acceptance_rate = marketing_df[col].mean() * 100
        total_accepted = marketing_df[col].sum()
        print(f"{col}: Acceptance Rate = {acceptance_rate:.2f}%, Total Accepted = {total_accepted}")

if len(campaign_cols) >= 2:
    print("\nA/B Testing: Comparing Campaign 1 vs Campaign 2")
    
    if 'AcceptedCmp1' in marketing_df.columns and 'AcceptedCmp2' in marketing_df.columns:
        group_a = marketing_df['AcceptedCmp1']
        group_b = marketing_df['AcceptedCmp2']
    else:
        group_a = marketing_df[campaign_cols[0]]
        group_b = marketing_df[campaign_cols[1]]
    
    mean_a = group_a.mean()
    mean_b = group_b.mean()
    
    t_stat, p_value = stats.ttest_ind(group_a, group_b)
    
    print(f"Campaign A Mean: {mean_a:.4f}")
    print(f"Campaign B Mean: {mean_b:.4f}")
    print(f"T-Statistic: {t_stat:.4f}")
    print(f"P-Value: {p_value:.4f}")
    
    if p_value < 0.05:
        print("Result: Statistically significant difference between campaigns")
    else:
        print("Result: No statistically significant difference between campaigns")

if 'Response' in marketing_df.columns and 'Income' in marketing_df.columns:
    responders = marketing_df[marketing_df['Response'] == 1]['Income']
    non_responders = marketing_df[marketing_df['Response'] == 0]['Income']
    
    print("\nResponse Analysis:")
    print(f"Average Income of Responders: {responders.mean():.2f}")
    print(f"Average Income of Non-Responders: {non_responders.mean():.2f}")
    
    t_stat_resp, p_value_resp = stats.ttest_ind(responders, non_responders)
    print(f"T-Statistic: {t_stat_resp:.4f}")
    print(f"P-Value: {p_value_resp:.4f}")

fig, axes = plt.subplots(2, 3, figsize=(18, 12))

if len(campaign_cols) > 0:
    axes[0, 0].bar(campaign_cols, marketing_df[campaign_cols].mean())
    axes[0, 0].set_title('Campaign Acceptance Rates')
    axes[0, 0].tick_params(axis='x', rotation=45)
    axes[0, 0].set_ylabel('Acceptance Rate')

if len(product_cols) > 0:
    axes[0, 1].bar(product_cols, marketing_df[product_cols].mean())
    axes[0, 1].set_title('Average Spending by Category')
    axes[0, 1].tick_params(axis='x', rotation=45)
    axes[0, 1].set_ylabel('Average Spending')

segment_counts = marketing_df['Customer_Segment'].value_counts()
axes[0, 2].pie(segment_counts, labels=segment_counts.index, autopct='%1.1f%%')
axes[0, 2].set_title('Customer Segment Distribution')

if 'Income' in marketing_df.columns and 'Customer_Segment' in marketing_df.columns:
    marketing_df.boxplot(column='Income', by='Customer_Segment', ax=axes[1, 0])
    axes[1, 0].set_title('Income by Customer Segment')
    axes[1, 0].set_xlabel('Segment')

if 'Total_Spending' in marketing_df.columns and 'Customer_Segment' in marketing_df.columns:
    marketing_df.boxplot(column='Total_Spending', by='Customer_Segment', ax=axes[1, 1])
    axes[1, 1].set_title('Spending by Customer Segment')
    axes[1, 1].set_xlabel('Segment')

if len(campaign_cols) > 0 and 'Customer_Segment' in marketing_df.columns:
    segment_campaign = marketing_df.groupby('Customer_Segment')[campaign_cols].mean()
    segment_campaign.plot(kind='bar', ax=axes[1, 2])
    axes[1, 2].set_title('Campaign Response by Segment')
    axes[1, 2].legend(bbox_to_anchor=(1.05, 1), loc='upper left')
    axes[1, 2].tick_params(axis='x', rotation=0)

plt.suptitle('Marketing Campaign Dashboard', fontsize=14, y=1.02)
plt.tight_layout()
plt.show()

print("\n" + "="*50)
print("BUSINESS INSIGHTS")
print("="*50)

print("\n1. CHURN PREDICTION INSIGHTS:")
print(f"   - Model Accuracy: {accuracy_score(y_test_churn, rf_preds):.2%}")
print(f"   - Top 3 Churn Drivers: {', '.join(feature_importance['Feature'].head(3).tolist())}")

print("\n2. CREDIT RISK INSIGHTS:")
print(f"   - Model AUC: {roc_auc_score(y_test_credit, rf_credit_probs):.4f}")
print(f"   - High Risk Applicants: {(credit_df['Risk_Category'] == 'High Risk').sum()}")
print(f"   - Low Risk Applicants: {(credit_df['Risk_Category'] == 'Low Risk').sum()}")

print("\n3. MARKETING INSIGHTS:")
if len(campaign_cols) > 0:
    best_campaign = marketing_df[campaign_cols].mean().idxmax()
    best_rate = marketing_df[campaign_cols].mean().max() * 100
    print(f"   - Best Performing Campaign: {best_campaign} ({best_rate:.2f}% acceptance)")
print(f"   - Number of Customer Segments: {marketing_df['Customer_Segment'].nunique()}")
if 'Total_Spending' in marketing_df.columns:
    highest_spending_segment = marketing_df.groupby('Customer_Segment')['Total_Spending'].mean().idxmax()
    print(f"   - Highest Spending Segment: Segment {highest_spending_segment}")
```
